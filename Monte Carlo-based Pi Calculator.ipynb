```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=100_000
inside=[]
for i in range(N):
    x=np.random.uniform(-1,1)
    y=np.random.uniform(-1,1)
    if np.sqrt(x**2+y**2)<1:
        inside.append((x,y))
```

```python
plt.figure(figsize=(6,6))
plt.scatter([x[0] for x in inside],[x[1] for x in inside], marker='.',alpha=.5)
```

```python
4*len(inside)/float(N)
```
